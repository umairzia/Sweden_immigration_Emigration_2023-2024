```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort and get top 10 by 2024 emigration numbers
top10_emigration = emigration_df.sort_values(by='2024 Total', ascending=False).head(10)

# Plot
plt.figure(figsize=(8, 6))
sns.barplot(
    x='2024 Total',
    y='Country of Birth',
    data=top10_emigration,
    hue='Country of Birth',
    dodge=False,
    palette='Reds_r',
    legend=False
)
plt.title('Top 10 Emigration Countries from Sweden (2024)')
plt.xlabel('Number of Emigrants')
plt.ylabel('Country of Birth')
plt.tight_layout()
plt.show()
```

```python
# Merge on Country of Birth
merged_df = pd.merge(
    emigration_df[['Country of Birth', '2024 Total']],
    immigration_df[['Country of Birth', '2024 Total']],
    on='Country of Birth',
    suffixes=('_Emigration', '_Immigration')
)
```

```python
# 4. Emigration vs Immigration (scatter)

plt.figure(figsize=(10, 6))

scatter = sns.scatterplot(
    x='2024 Total_Emigration',
    y='2024 Total_Immigration',
    data=merged_df,
    hue='Country of Birth',
    palette='tab10',
    s=100
)

plt.title('Emigration vs Immigration by Country (Sweden, 2024)')
plt.xlabel('Emigration (2024 Total)')
plt.ylabel('Immigration (2024 Total)')
plt.grid(True)

# Move legend outside the plot
plt.legend(
    title='Country of Birth',
    bbox_to_anchor=(1.05, 1),  # place legend outside the plot
    loc='upper left'
)

plt.tight_layout()
plt.show()
```

```python
# Prepare data for total immigration (wide to long format)
immigration_total_long = pd.melt(
    immigration_df,
    id_vars=['Country of Birth'],
    value_vars=['2023 Total', '2024 Total'],
    var_name='Year',
    value_name='Immigrants'
)
# Clean 'Year' column to keep just 2023 or 2024
immigration_total_long['Year'] = immigration_total_long['Year'].str.extract(r'(\d{4})')
# Sort by immigration volume (2024) to improve readability
top_countries = immigration_df.sort_values(by='2024 Total', ascending=False)['Country of Birth'].head(10)
immigration_total_long = immigration_total_long[immigration_total_long['Country of Birth'].isin(top_countries)]
```

```python
# Plot: Total Immigration Change (2023 vs 2024)
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=immigration_total_long,
    x='Year',
    y='Immigrants',
    hue='Country of Birth',
    marker='o',
    palette='tab10'
)

plt.title('Immigration to Sweden (Top 10 Countries): 2023 vs 2024')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.grid(True)
plt.legend(title='Country of Birth', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Melt gender-specific columns
immigration_gender_long = pd.melt(
    immigration_df,
    id_vars=['Country of Birth'],
    value_vars=['2023 Women', '2024 Women', '2023 Men', '2024 Men'],
    var_name='Year_Gender',
    value_name='Immigrants'
)

# Split 'Year_Gender' into two columns: Year and Gender
immigration_gender_long[['Year', 'Gender']] = immigration_gender_long['Year_Gender'].str.extract(r'(\d{4})\s*(Women|Men)')
immigration_gender_long.drop(columns='Year_Gender', inplace=True)

# Filter only top 10 countries
immigration_gender_long = immigration_gender_long[immigration_gender_long['Country of Birth'].isin(top_countries)]

# Facet vertically: Men on top, Women on bottom
g = sns.FacetGrid(
    immigration_gender_long,
    row='Gender',        # stack plots vertically
    sharex=True,
    height=4,
    aspect=1.5
)

g.map_dataframe(
    sns.lineplot,
    x='Immigrants',
    y='Year',
    hue='Country of Birth',
    marker='o',
    palette='tab10'
)

g.set_axis_labels("Number of Immigrants", "Year")
g.add_legend(title='Country of Birth', bbox_to_anchor=(1.05, 1), loc='upper left')
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('Immigration to Sweden by Gender (Top 10 Countries): 2023 vs 2024', fontsize=14)

plt.tight_layout()
plt.show()
```
